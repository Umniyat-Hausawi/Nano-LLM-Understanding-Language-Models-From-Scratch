# 🔢 Encoding and Decoding

After building the vocabulary, we need to convert characters into numerical IDs.

This step is called encoding.

Neural networks cannot process raw text directly, so each character must be represented as a number before being passed into the model.

We also create the reverse process, called decoding, which converts numerical IDs back into readable text.

At this stage, we will:

- Create a character-to-index mapping
- Create an index-to-character mapping
- Encode text into numerical IDs
- Decode numerical IDs back into text
- Verify that encoding and decoding work correctly

````python
# Create character-to-index and index-to-character mappings
char_to_idx = {char: idx for idx, char in enumerate(chars)}
idx_to_char = {idx: char for idx, char in enumerate(chars)}

# Define encoding and decoding functions
def encode(text):
    return [char_to_idx[char] for char in text]

def decode(indices):
    return ''.join(idx_to_char[idx] for idx in indices)

# Test encoding and decoding
sample_text = "Lily likes cats."

encoded_sample = encode(sample_text)
decoded_sample = decode(encoded_sample)

print("Original text:")
print(sample_text)

print("\nEncoded text:")
print(encoded_sample)

print("\nDecoded text:")
print(decoded_sample)
````

# 🧠 Creating Input and Target Sequences

A language model learns by predicting the next token.

After converting text into numerical IDs, we need to create training examples where the input sequence contains previous tokens, and the target sequence contains the same text shifted one step forward.

This teaches the model:

> Given this context, predict the next character.

For example:

Input:
L i l

Target:
i l y

This means the model learns:

- L → i
- L i → l
- L i l → y

At this stage, we will:

- Convert the full training text into encoded IDs
- Define a context window using `block_size`
- Create input-target pairs
- Understand how next-token prediction is prepared for training

# 🛡️ Handling Unknown Characters

During validation encoding, we found that the validation set contains a character that does not exist in the training vocabulary.

This is called an out-of-vocabulary token.

In a production-style machine learning workflow, we should not build the vocabulary using validation data, because validation data should remain unseen during training preparation.

To handle unseen characters properly, we add a special `<UNK>` token to the training vocabulary.

Any character that does not exist in the training vocabulary will be mapped to `<UNK>` during encoding.

````python
# Add a special token for unknown characters
UNK_TOKEN = "<UNK>"

# Rebuild vocabulary from training text only
chars = sorted(list(set(train_text)))

# Add UNK token to the vocabulary
chars.append(UNK_TOKEN)

# Recalculate vocabulary size
vocab_size = len(chars)

# Rebuild mappings
char_to_idx = {char: idx for idx, char in enumerate(chars)}
idx_to_char = {idx: char for idx, char in enumerate(chars)}

print("Vocabulary size with <UNK>:", vocab_size)
print("UNK token ID:", char_to_idx[UNK_TOKEN])
````

````python
#  encoding function to handle unknown characters
def encode(text):
    return [char_to_idx.get(char, char_to_idx[UNK_TOKEN]) for char in text]

def decode(indices):
    return ''.join(idx_to_char[idx] for idx in indices)

# Test encoding validation text again
train_data = encode(train_text)
val_data = encode(val_text)

print("Number of training tokens:", len(train_data))
print("Number of validation tokens:", len(val_data))

print("\nDecoded first 100 training tokens:")
print(decode(train_data[:100]))

print("\nDecoded first 100 validation tokens:")
print(decode(val_data[:100]))
````

# 🏗️ Building the Training Dataset

Instead of using a single training example, we create many input-target pairs using a sliding window approach.

The input sequence (`x`) contains previous characters, while the target sequence (`y`) contains the same text shifted by one position.

This allows the model to learn next-character prediction across the entire dataset.

At this stage, we will:

- Create a dataset function
- Generate training examples dynamically
- Inspect sample input-target pairs

````python
import torch

# Convert encoded text into PyTorch tensors
train_data = torch.tensor(train_data, dtype=torch.long)
val_data = torch.tensor(val_data, dtype=torch.long)

# Function to generate one batch
def get_batch(data, block_size=64, batch_size=4):

    # Random starting positions
    ix = torch.randint(len(data) - block_size, (batch_size,))

    # Input sequences
    x = torch.stack([data[i:i+block_size] for i in ix])

    # Target sequences (shifted by one token)
    y = torch.stack([data[i+1:i+block_size+1] for i in ix])

    return x, y


# Test batch generation
x_batch, y_batch = get_batch(train_data)

print("Input batch shape:", x_batch.shape)
print("Target batch shape:", y_batch.shape)
````

# 🧠 Building the Bigram Language Model

We are now ready to build our first language model.

A Bigram Language Model learns to predict the next token based only on the current token.

This is a very simple language model, but it helps us understand the fundamental mechanics behind next-token prediction.

At this stage, we will:

- Build a simple neural network
- Predict the next token
- Generate probability distributions
- Understand how language models learn patterns

````python
import torch.nn as nn
import torch.nn.functional as F


class BigramLanguageModel(nn.Module):

    def __init__(self, vocab_size):
        super().__init__()

        # Lookup table: token → next-token probabilities
        self.token_embedding_table = nn.Embedding(
            vocab_size,
            vocab_size
        )

    def forward(self, idx, targets=None):

        # Get logits
        logits = self.token_embedding_table(idx)

        # Calculate loss if targets exist
        if targets is None:
            loss = None
        else:
            B, T, C = logits.shape

            logits = logits.view(B * T, C)
            targets = targets.view(B * T)

            loss = F.cross_entropy(logits, targets)

        return logits, loss
````

````python
# Create model instance
model = BigramLanguageModel(vocab_size)

# Get one batch
xb, yb = get_batch(train_data)

# Forward pass
logits, loss = model(xb, yb)

print("Logits shape:", logits.shape)
print("Initial loss:", loss.item())
````

# 🏋️ Training the Bigram Model

Now we will train our first language model.

During training, the model repeatedly:

1. Receives input sequences
2. Predicts the next token
3. Calculates prediction error (loss)
4. Computes gradients through backpropagation
5. Updates weights to improve future predictions

The goal is to gradually reduce the loss and improve next-token prediction performance.

````python
# Create optimizer
optimizer = torch.optim.AdamW(
    model.parameters(),
    lr=1e-3
)

# Training loop
max_iters = 3000

for step in range(max_iters):

    # Get batch
    xb, yb = get_batch(train_data)

    # Forward pass
    logits, loss = model(xb, yb)

    # Reset gradients
    optimizer.zero_grad(set_to_none=True)

    # Backward pass
    loss.backward()

    # Update weights
    optimizer.step()

    # Print progress
    if step % 300 == 0:
        print(f"Step {step}, Loss: {loss.item():.4f}")
````

# ✍️ Text Generation with the Bigram Model

After training the Bigram Language Model, we can use it to generate new text.

Text generation happens autoregressively:

1. Start with an initial token
2. Predict the next token
3. Add the predicted token to the sequence
4. Use the updated sequence to predict the next token
5. Repeat the process

This helps us evaluate whether the model has learned basic language patterns from the training data.

````python
# Add text generation method to the Bigram model
def generate_text(model, start_text, max_new_tokens=300):
    model.eval()

    # Encode the starting text
    context = torch.tensor([encode(start_text)], dtype=torch.long)

    # Generate tokens one by one
    for _ in range(max_new_tokens):

        # Get predictions
        logits, loss = model(context)

        # Focus only on the last token prediction
        logits = logits[:, -1, :]

        # Convert logits to probabilities
        probs = F.softmax(logits, dim=-1)

        # Sample next token
        next_token = torch.multinomial(probs, num_samples=1)

        # Append predicted token to context
        context = torch.cat((context, next_token), dim=1)

    # Decode generated IDs back to text
    return decode(context[0].tolist())


# Generate sample text
generated_text = generate_text(
    model,
    start_text="Once upon a time",
    max_new_tokens=500
)

print(generated_text)
````

# 📉 Evaluating the Bigram Model

After training the Bigram model, we need to evaluate its performance on both training and validation data.

Training loss shows how well the model learned from the data it was trained on.

Validation loss shows how well the model performs on unseen data.

If training loss is much lower than validation loss, this may indicate overfitting.

````python
@torch.no_grad()
def estimate_loss(model, train_data, val_data, eval_iters=100):
    model.eval()
    losses = {}

    for split, data in [("train", train_data), ("val", val_data)]:
        split_losses = []

        for _ in range(eval_iters):
            xb, yb = get_batch(data)
            logits, loss = model(xb, yb)
            split_losses.append(loss.item())

        losses[split] = sum(split_losses) / len(split_losses)

    model.train()
    return losses


# Estimate train and validation loss
losses = estimate_loss(model, train_data, val_data)

print("Train loss:", losses["train"])
print("Validation loss:", losses["val"])
````

# 🔧 Improving the Bigram Model

Although the Bigram model is simple, we can try improving it before moving to a stronger architecture.

We will improve it from two directions:

1. Training improvements:
   - Increase batch size
   - Train for more iterations

2. Generation improvements:
   - Add temperature sampling
   - Add top-k sampling

This helps us understand the difference between improving the training process and improving the text generation strategy.

````python
# Updated batch function with larger batch size
def get_batch(data, block_size=64, batch_size=32):
    ix = torch.randint(len(data) - block_size, (batch_size,))

    x = torch.stack([data[i:i+block_size] for i in ix])
    y = torch.stack([data[i+1:i+block_size+1] for i in ix])

    return x, y


# Continue training the same Bigram model
optimizer = torch.optim.AdamW(
    model.parameters(),
    lr=1e-3
)

max_iters = 5000

for step in range(max_iters):

    xb, yb = get_batch(train_data)

    logits, loss = model(xb, yb)

    optimizer.zero_grad(set_to_none=True)
    loss.backward()
    optimizer.step()

    if step % 500 == 0:
        print(f"Step {step}, Loss: {loss.item():.4f}")
````

````python
def generate_text_improved(model, start_text, max_new_tokens=500, temperature=0.8, top_k=20):
    model.eval()

    context = torch.tensor([encode(start_text)], dtype=torch.long)

    for _ in range(max_new_tokens):

        logits, loss = model(context)

        # Use only the last token prediction
        logits = logits[:, -1, :]

        # Apply temperature
        logits = logits / temperature

        # Apply top-k filtering
        if top_k is not None:
            values, indices = torch.topk(logits, top_k)
            filtered_logits = torch.full_like(logits, float("-inf"))
            filtered_logits.scatter_(1, indices, values)
            logits = filtered_logits

        # Convert logits to probabilities
        probs = F.softmax(logits, dim=-1)

        # Sample next token
        next_token = torch.multinomial(probs, num_samples=1)

        # Add predicted token
        context = torch.cat((context, next_token), dim=1)

    return decode(context[0].tolist())


# Generate improved sample
generated_text = generate_text_improved(
    model,
    start_text="Once upon a time",
    max_new_tokens=500,
    temperature=0.7,
    top_k=15
)

print(generated_text)
````

## Bigram Model Limitation

Although the Bigram model improved after training, its generated text still lacks coherent meaning.

The model learned basic character-level patterns such as common letters, spaces, and short word-like structures. However, it cannot understand long-range context because it only uses the current token to predict the next token.

As a result, the generated text may look similar to English, but many words are meaningless and the sentences are not logically connected.

This shows that improving training alone is not enough. To generate more coherent text, we need a model that can use longer context and capture relationships between tokens.

# 🧠 Context-Aware Character Model

The Bigram model could only use the current token to predict the next token.

In this stage, we improve the model by allowing it to see a larger context window.

Instead of looking at only one token, the model will use the previous 64 tokens to predict the next token.

This helps the model capture longer patterns and generate more coherent text.

# 🧠 Building a Context-Aware Character Model

The Bigram model could only use one token to predict the next token.

In this stage, we improve the model by allowing it to use a larger context window.

Instead of looking at only one token, the model will use the previous 64 tokens to predict the next token.

This helps the model learn longer patterns and generate more coherent text.

````python
class ContextAwareLanguageModel(nn.Module):

    def __init__(self, vocab_size, embedding_dim=64, block_size=64):
        super().__init__()

        self.block_size = block_size

        # Token embeddings
        self.token_embedding_table = nn.Embedding(
            vocab_size,
            embedding_dim
        )

        # Linear layer for prediction
        self.language_model_head = nn.Linear(
            embedding_dim * block_size,
            vocab_size
        )

    def forward(self, idx, targets=None):

        # Get token embeddings
        x = self.token_embedding_table(idx)

        # Flatten context
        B, T, C = x.shape
        x = x.view(B, T * C)

        # Predict next token
        logits = self.language_model_head(x)

        # Calculate loss
        if targets is None:
            loss = None
        else:
            targets = targets[:, -1]
            loss = F.cross_entropy(logits, targets)

        return logits, loss
````

````python
# Create model
context_model = ContextAwareLanguageModel(
    vocab_size=vocab_size,
    embedding_dim=64,
    block_size=64
)

# Test one batch
xb, yb = get_batch(train_data)

logits, loss = context_model(xb, yb)

print("Logits shape:", logits.shape)
print("Initial loss:", loss.item())
````

````python
optimizer = torch.optim.AdamW(
    context_model.parameters(),
    lr=1e-3
)

max_iters = 5000

for step in range(max_iters):

    xb, yb = get_batch(train_data)

    logits, loss = context_model(xb, yb)

    optimizer.zero_grad(set_to_none=True)

    loss.backward()

    optimizer.step()

    if step % 500 == 0:
        print(f"Step {step}, Loss: {loss.item():.4f}")
````

````python
losses = estimate_loss(
    context_model,
    train_data,
    val_data
)

print("Train loss:", losses["train"])
print("Validation loss:", losses["val"])
````

````python
def generate_text_context(
    model,
    start_text,
    max_new_tokens=500,
    temperature=0.8,
    top_k=20
):

    model.eval()

    # Encode start text
    context = torch.tensor(
        [encode(start_text)],
        dtype=torch.long
    )

    for _ in range(max_new_tokens):

        # Keep only last block_size tokens
        context_crop = context[:, -64:]

        # Pad if shorter than block_size
        if context_crop.shape[1] < 64:
            padding = torch.zeros(
                (1, 64 - context_crop.shape[1]),
                dtype=torch.long
            )

            context_crop = torch.cat(
                (padding, context_crop),
                dim=1
            )

        # Forward pass
        logits, _ = model(context_crop)

        # Apply temperature
        logits = logits / temperature

        # Top-k filtering
        if top_k is not None:
            values, indices = torch.topk(logits, top_k)

            filtered_logits = torch.full_like(
                logits,
                float("-inf")
            )

            filtered_logits.scatter_(
                1,
                indices,
                values
            )

            logits = filtered_logits

        # Convert to probabilities
        probs = F.softmax(logits, dim=-1)

        # Sample next token
        next_token = torch.multinomial(
            probs,
            num_samples=1
        )

        # Add predicted token
        context = torch.cat(
            (context, next_token),
            dim=1
        )

    return decode(context[0].tolist())


# Generate text
generated_text = generate_text_context(
    context_model,
    start_text="Once upon a time",
    max_new_tokens=500,
    temperature=0.7,
    top_k=15
)

print(generated_text)
````

# 🧩 Subword Tokenization

Character-level tokenization improved context understanding, but text generation still produced many meaningless words.

To improve text quality, we move to subword tokenization.

Instead of splitting text into single characters or full words, subword tokenization breaks text into meaningful pieces.

For example:

playing → play + ing  
wonderful → wonder + ful  
smartphone → smart + phone

This approach balances vocabulary size, flexibility, and text quality.

Modern language models such as GPT use subword tokenization because it handles unknown words better while preserving meaningful language structure.

# 📉 Evaluate and Generate Text

After training the subword-based model, we evaluate it using training and validation loss.

Then we generate new text to qualitatively compare the output with the previous character-level models.

This helps us understand the practical impact of tokenization choice on language generation quality.

````python
def generate_text_subword(
    model,
    start_text="Once upon a time",
    max_new_tokens=300,
    temperature=0.8,
    top_k=20
):
    model.eval()

    context = torch.tensor(
        [encode_subword(start_text)],
        dtype=torch.long
    )

    with torch.no_grad():
        for _ in range(max_new_tokens):

            # Keep last block_size tokens
            context_crop = context[:, -64:]

            # Pad if context is shorter than block_size
            if context_crop.shape[1] < 64:
                padding_length = 64 - context_crop.shape[1]

                padding = torch.zeros(
                    (1, padding_length),
                    dtype=torch.long
                )

                context_crop = torch.cat(
                    (padding, context_crop),
                    dim=1
                )

            logits, _ = model(context_crop)

            logits = logits / temperature

            if top_k is not None:
                values, indices = torch.topk(logits, top_k)
                probs = F.softmax(values, dim=-1)

                sampled_idx = torch.multinomial(
                    probs,
                    num_samples=1
                )

                next_token = indices.gather(
                    -1,
                    sampled_idx
                )
            else:
                probs = F.softmax(logits, dim=-1)
                next_token = torch.multinomial(
                    probs,
                    num_samples=1
                )

            context = torch.cat(
                (context, next_token),
                dim=1
            )

    return decode_subword(context[0].tolist())
````

# 🧠 Building Token and Positional Embeddings

Before the Transformer can understand language, it first needs a way to represent tokens numerically.

Each token is converted into a dense vector representation called an **Embedding**.

However, token meaning alone is not enough.

The model must also understand **word order**.

For example:

> "dog bites man"

is different from:

> "man bites dog"

Even though both sentences contain the same words.

To solve this problem, Transformer models combine:

### 1. Token Embeddings
Represent the meaning of each token.

### 2. Positional Embeddings
Represent the position of each token in the sequence.

The final token representation becomes:

```text
Token Meaning + Position Information
```

This allows the model to understand both **what the token means** and **where it appears in the sentence**.

````python
class TransformerLanguageModel(nn.Module):

    def __init__(
        self,
        vocab_size,
        embedding_dim=64,
        block_size=64
    ):
        super().__init__()

        self.block_size = block_size

        # Token embedding
        self.token_embedding_table = nn.Embedding(
            vocab_size,
            embedding_dim
        )

        # Positional embedding
        self.position_embedding_table = nn.Embedding(
            block_size,
            embedding_dim
        )

    def forward(self, idx):

        B, T = idx.shape

        # Token embeddings
        token_embeddings = self.token_embedding_table(idx)

        # Position embeddings
        positions = torch.arange(T)

        position_embeddings = (
            self.position_embedding_table(
                positions
            )
        )

        # Combine both
        x = token_embeddings + position_embeddings

        return x
````

# 🔎 Self-Attention

Self-Attention is the core idea behind Transformer models.

Instead of treating all previous tokens equally, self-attention allows each token to look at other tokens in the context and decide which ones are more important.

For example, in the sentence:

> The bird flew because it was scared.

The token `it` should pay more attention to `bird` than to unrelated words.

This helps the model capture relationships between tokens and understand context more effectively.

In a decoder-only language model, we also use causal masking so the model cannot look at future tokens while predicting the next token.

````python
class Head(nn.Module):

    def __init__(self, head_size):

        super().__init__()

        # Key
        self.key = nn.Linear(
            embedding_dim,
            head_size,
            bias=False
        )

        # Query
        self.query = nn.Linear(
            embedding_dim,
            head_size,
            bias=False
        )

        # Value
        self.value = nn.Linear(
            embedding_dim,
            head_size,
            bias=False
        )

        # Causal mask
        self.register_buffer(
            "tril",
            torch.tril(
                torch.ones(
                    block_size,
                    block_size
                )
            )
        )

    def forward(self, x):

        B, T, C = x.shape

        # Create K, Q, V
        K = self.key(x)
        Q = self.query(x)
        V = self.value(x)

        # Attention scores
        weights = (
            Q @ K.transpose(-2, -1)
        ) / (K.shape[-1] ** 0.5)

        # Apply causal masking
        weights = weights.masked_fill(
            self.tril[:T, :T] == 0,
            float("-inf")
        )

        # Convert to probabilities
        weights = F.softmax(
            weights,
            dim=-1
        )

        # Weighted aggregation
        out = weights @ V

        return out
````

# 🧠 Multi-Head Attention

A single attention head can only learn one type of relationship.

To improve understanding, Transformers use multiple attention heads in parallel.

Each head learns different patterns in the data.

For example:

- One head may focus on grammar
- Another on long-range context
- Another on word relationships

The outputs from all heads are combined together to create a richer representation.

This is one of the major breakthroughs behind Transformer models.

````python
class MultiHeadAttention(nn.Module):

    def __init__(
        self,
        num_heads,
        head_size
    ):

        super().__init__()

        self.heads = nn.ModuleList(
            [
                Head(head_size)
                for _ in range(num_heads)
            ]
        )

        self.proj = nn.Linear(
            num_heads * head_size,
            embedding_dim
        )

    def forward(self, x):

        out = torch.cat(
            [h(x) for h in self.heads],
            dim=-1
        )

        out = self.proj(out)

        return out
````

````python
multi_head = MultiHeadAttention(
    num_heads=4,
    head_size=16
)

xb, yb = get_batch(train_data)

x = model(xb)

out = multi_head(x)

print(out.shape)
````

# 🧠 Feed Forward Network

After attention determines which tokens are important, the model still needs a way to process this information.

This is the role of the **Feed Forward Network (FFN)**.

The FFN acts like a small neural network applied independently to every token representation.

Its goal is to:

- Learn more complex patterns
- Refine token representations
- Improve understanding after attention

In Transformer models, the sequence typically becomes:

```text
Attention → Feed Forward
```

This combination helps the model both **focus** and **reason** about the context.

````python
class FeedForward(nn.Module):

    def __init__(self, embedding_dim):

        super().__init__()

        self.net = nn.Sequential(

            nn.Linear(
                embedding_dim,
                4 * embedding_dim
            ),

            nn.ReLU(),

            nn.Linear(
                4 * embedding_dim,
                embedding_dim
            )
        )

    def forward(self, x):

        return self.net(x)
````

````python
ffn = FeedForward(
    embedding_dim=64
)

xb, yb = get_batch(train_data)

x = model(xb)

x = multi_head(x)

out = ffn(x)

print(out.shape)
````

# 🧱 Transformer Block

A Transformer Block combines the two most important components of the Transformer architecture:

1. Multi-Head Attention  
2. Feed Forward Network

The attention mechanism helps the model decide which tokens are important.

The feed forward network then processes this information to learn more complex patterns.

In modern LLMs such as GPT, multiple Transformer Blocks are stacked together to build deep language understanding.

This block acts as the core building unit of Transformer-based language models.

````python
class TransformerBlock(nn.Module):

    def __init__(
        self,
        embedding_dim,
        num_heads
    ):

        super().__init__()

        head_size = (
            embedding_dim
            // num_heads
        )

        self.attention = (
            MultiHeadAttention(
                num_heads,
                head_size
            )
        )

        self.ffn = FeedForward(
            embedding_dim
        )

    def forward(self, x):

        x = self.attention(x)

        x = self.ffn(x)

        return x
````

````python
block = TransformerBlock(
    embedding_dim=64,
    num_heads=4
)

xb, yb = get_batch(train_data)

x = model(xb)

out = block(x)

print(out.shape)
````

# ⚙️ Residual Connections and Layer Normalization

Real Transformer models do not simply pass data through attention and feed-forward layers directly.

They also use two important techniques:

### 1. Residual Connections
Residual connections allow the model to keep the original information while adding newly learned information.

This helps gradients flow better during training and makes deeper models easier to train.

### 2. Layer Normalization
Layer normalization stabilizes training by normalizing token representations.

Together, these techniques make Transformer training more stable and effective.

````python
class TransformerBlock(nn.Module):

    def __init__(self, embedding_dim, num_heads):
        super().__init__()

        head_size = embedding_dim // num_heads

        self.attention = MultiHeadAttention(
            num_heads=num_heads,
            head_size=head_size
        )

        self.ffn = FeedForward(embedding_dim)

        self.ln1 = nn.LayerNorm(embedding_dim)
        self.ln2 = nn.LayerNorm(embedding_dim)

    def forward(self, x):

        # Residual connection around attention
        x = x + self.attention(self.ln1(x))

        # Residual connection around feed-forward network
        x = x + self.ffn(self.ln2(x))

        return x
````

````python
block = TransformerBlock(
    embedding_dim=64,
    num_heads=4
)

xb, yb = get_batch(train_data)

x = model(xb)

out = block(x)

print(out.shape)
````

# 🧱 Building the Full Transformer Model

So far, we built the main components of a Transformer:

- Token Embeddings
- Positional Embeddings
- Self-Attention
- Multi-Head Attention
- Feed Forward Networks
- Residual Connections
- Layer Normalization

Now we combine everything into a complete Transformer model.

Instead of using a single block, we stack multiple Transformer blocks to allow deeper understanding of language patterns.

This is one of the key ideas behind modern LLMs such as GPT.

````python
class GPTLanguageModel(nn.Module):

    def __init__(
        self,
        vocab_size,
        embedding_dim=64,
        block_size=64,
        num_heads=4,
        num_layers=2
    ):
        super().__init__()

        self.block_size = block_size

        # Token embeddings
        self.token_embedding_table = nn.Embedding(
            vocab_size,
            embedding_dim
        )

        # Position embeddings
        self.position_embedding_table = nn.Embedding(
            block_size,
            embedding_dim
        )

        # Stack Transformer blocks
        self.blocks = nn.Sequential(
            *[
                TransformerBlock(
                    embedding_dim=embedding_dim,
                    num_heads=num_heads
                )
                for _ in range(num_layers)
            ]
        )

        # Final normalization
        self.ln_f = nn.LayerNorm(embedding_dim)

        # Final prediction layer
        self.lm_head = nn.Linear(
            embedding_dim,
            vocab_size
        )

    def forward(self, idx, targets=None):

        B, T = idx.shape

        # Token embeddings
        tok_emb = self.token_embedding_table(idx)

        # Position embeddings
        pos_emb = self.position_embedding_table(
            torch.arange(T, device=idx.device)
        )

        # Combine
        x = tok_emb + pos_emb

        # Transformer blocks
        x = self.blocks(x)

        # Final normalization
        x = self.ln_f(x)

        # Final logits
        logits = self.lm_head(x)

        loss = None

        if targets is not None:

            B, T, C = logits.shape

            logits = logits.view(B * T, C)
            targets = targets.view(B * T)

            loss = F.cross_entropy(
                logits,
                targets
            )

        return logits, loss
````

````python
model = GPTLanguageModel(
    vocab_size=vocab_size,
    embedding_dim=64,
    block_size=64,
    num_heads=4,
    num_layers=2
)

xb, yb = get_batch(train_data)

logits, loss = model(xb, yb)

print("Logits shape:", logits.shape)
print("Initial loss:", loss.item())
````

# ✨ Text Generation

Now we test the Transformer by generating new text.

The model predicts one token at a time based on previous context.

We use:

- Temperature → controls randomness
- Top-K Sampling → limits predictions to the most likely tokens

This helps produce more coherent and meaningful text generation.

````python
def generate_text_transformer(
    model,
    start_text="Once upon a time",
    max_new_tokens=300,
    temperature=0.8,
    top_k=20
):

    model.eval()

    encoded = encode_subword(start_text)

    context = torch.tensor(
        [encoded],
        dtype=torch.long
    )

    for _ in range(max_new_tokens):

        context_crop = context[:, -64:]

        logits, _ = model(context_crop)

        logits = logits[:, -1, :]

        logits = logits / temperature

        values, indices = torch.topk(
            logits,
            top_k
        )

        probs = F.softmax(
            values,
            dim=-1
        )

        next_idx = torch.multinomial(
            probs,
            num_samples=1
        )

        next_token = torch.gather(
            indices,
            -1,
            next_idx
        )

        context = torch.cat(
            [context, next_token],
            dim=1
        )

    generated_ids = context[0].tolist()

    return decode_subword(
        generated_ids
    )
````

````python
generated_text = generate_text_transformer(
    model,
    start_text="Once upon a time",
    max_new_tokens=300,
    temperature=0.8,
    top_k=20
)

print(generated_text)
````
